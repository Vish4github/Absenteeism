```python
month_list=[]
```

```python
for i in range(abs_data_check.shape[0]):
    month_list.append(abs_data_check['Date'][i].month)
```

```python
#month_list
abs_data_check['Month']=month_list
```

```python
#abs_data_check.head()
```

```python
def day_of_week(date_value):
    return date_value.weekday()
```

```python
abs_data_check['Day of week']=abs_data_check['Date'].apply(day_of_week)
```

```python
abs_data_check=abs_data_check.drop('Date',axis=1)
```
