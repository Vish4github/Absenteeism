## Standardizing Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(unscaled_inputs) # calculates mean and std dev and stores it in scaler, so standardization info would always be in scaler now which can be reused as new data comes in
```

```python
scaled_inputs = scaler.transform(unscaled_inputs) # transform to actually use the scaled info
```

```python
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split
```

## Split

```python
train_test_split(scaled_inputs,targets)
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,targets,train_size=0.8,random_state=20)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

### working of .score()

```python
predicted =reg.predict(x_train)
```

```python
predicted == y_train
```

```python
np.sum(predicted == y_train)
```

```python
np.sum(predicted == y_train)/y_train.shape[0]
```

Getting the intercepts and coefficients

```python
reg.intercept_[0]
```

```python
reg.coef_
```

```python
column_values=unscaled_inputs.columns.values
```

```python
summary_table= pd.DataFrame(columns=['Feature name'],data=column_values)  ##creating the dataframe, defining col value and data
```

```python
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index=summary_table.index-1 # moving index by 1 so that we can add the intercept here
```

```python
summary_table.loc[0]=['Intercept',reg.intercept_[0]]
```

```python
summary_table= summary_table.sort_index()
```

```python
summary_table
```

## Interpreting Coefficients

```python
type(summary_table.Coefficient[0])
summary_table.Coefficient = np.float64(summary_table.Coefficient)
#summary_table['Odds_ratio']=map(float(),np.exp(summary_table.Coefficient))
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table.sort_values('Odds_ratio',ascending=False)
```
